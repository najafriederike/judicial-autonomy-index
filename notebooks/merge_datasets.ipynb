## Merge Judicial Autonomy and V-Dem Datasets

```python
import yaml

import pandas as pd
import numpy as np
```

```python
try:
    with open('../config.yaml') as file:
        config = yaml.safe_load(file)
except:
    print("Yaml file not found.")
```

### 1 Load Datasets (using clean datasets)

```python
# Load clean Judicial Autonomy Dataset

q1_cleaned = pd.read_csv(config['output_data']['q1_dataset'])
q1_cleaned
```

```python
# Load clean Judicial Autonomy index and macro indicators

q1_index_macro = pd.read_csv(config['output_data']['q1_output_index_macro'])
q1_index_macro
```

```python
# Load clean Judicial Autonomy Dataset index, macro indicators and micro indicators

q1_index_macro_micro = pd.read_csv(config['output_data']['q1_output_index_macro_micro'])
q1_index_macro_micro
```

```python
# Load clean V-Dem Dataset

vdem_cleaned = pd.read_csv(config['output_data']['vdem_episodes_of_regime_transformation_dataset_clean'])
vdem_cleaned
```

### 2 Merge Datasets

#### 2.1 Judicial Autonomy index and macro indicators subset & V-Dem Dataset

```python
# Merge datasets
q1_index_macro_vdem = pd.merge(q1_index_macro, vdem_cleaned, how='outer', on=['country', 'year'])

# Export dataset
q1_index_macro_vdem.to_csv(config['output_data']['q1_output_index_macro_vdem'], index=False)

# View dataset
q1_index_macro_vdem
```

#### 2.2 Judicial Autonomy index, macro indicators and micro indicators subset & V-Dem Dataset

```python
# Merge datasets
q1_index_macro_micro_vdem = pd.merge(q1_index_macro_micro, vdem_cleaned, how='outer', on=['country', 'year'])

# Export dataset
q1_index_macro_micro_vdem.to_csv(config['output_data']['q1_output_index_macro_micro_vdem'], index=False)

#View dataset
q1_index_macro_micro_vdem
```
